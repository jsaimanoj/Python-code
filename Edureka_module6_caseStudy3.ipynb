```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
```

# Company wants to decide whether to focus the effort on mobile app experience or its website. As a drastic measure it is also evaluating to shut down the website.

```python
x = data.iloc[:,3:7].values
x
```

```python
y = data.iloc[:,7].values
y
```

```python
x_std = sc_x.fit_transform(x)
x_std
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x_std,y, test_size = 0.25)
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
plt.axes([0,0,0.9,0.9])
plt.plot(y_pred)
plt.axes([1,0,0.9,0.9])
plt.plot(y_test)
```
